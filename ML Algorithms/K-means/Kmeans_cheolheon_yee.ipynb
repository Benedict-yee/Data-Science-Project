### Set toy dataset

```python
x = []  # data
k = 3  # hyper paramters
np.random.seed(2021)
x.extend(np.random.normal(loc=[0,0], scale=0.5, size=(100, 2)).tolist())
x.extend(np.random.normal(loc=[2,2], scale=0.5, size=(100, 2)).tolist())
x.extend(np.random.normal(loc=[-3,3], scale=0.5, size=(100, 2)).tolist())
x = np.array(x)

sns.scatterplot(x=x[:,0], y=x[:,1]);
```

### Build Kmeans algorithm and train model

```python
def distance(a, b):
    return sum((a - b) ** 2)/len(a)
```

```python
def group_center(g):
    g = np.array(g)
    return g.mean(axis=0)
```

```python
def clust(x, k, seed=2022, iter_num=100):
    logs = []
    np.random.seed(seed)
    centers = x[np.random.choice(len(x), size=k, replace=False)]
    for it in range(iter_num):
        group = {}
        for i in range(k):
            group[i] = []
        # find nearest center
        for row in x:
            temp = []
            for i in range(k):
                temp.append(distance(centers[i], row))
            group[np.argmin(temp)].append(row.tolist())

        # plot data store
        for i in range(k):
            group_temp = np.array(group[i])
            group_temp = np.c_[group_temp, np.full(len(group_temp), i)]
            if i == 0:
                grouped = group_temp
            else:
                grouped = np.append(grouped, group_temp, axis = 0)

        # update center
        centers_new = []
        for i in range(k):
            centers_new.append(group_center(group[i]).tolist())
        centers_new = np.array(centers_new)
        # if updated center == center, break

        if np.sum(centers - centers_new) == 0:
            break
        else:
            centers = centers_new
            logs.append(grouped)
    return grouped, logs, it  
```

### Apply toy dataset (K=3)

```python
grouped, logs, it = cluster(x, 3)
print(f'iter num:{it}')
```

```python
plt.figure(figsize=(15,8))
for i in range(it):
    plt.subplot(2, it//2+1, i+1)  # row, col, index
    df = pd.DataFrame(logs[i])
    df.columns = ['x1', 'x2', 'group']
    sns.scatterplot(data = df, x = 'x1', y = 'x2', hue = 'group').set_title(f'iter : {i}')
```

# Conclusions

#### From the result, I could find that K-means can easily separate clusters with simple algorithms. However, fashion MNIST dataset is not a good sample for K-means clustering because it contains a lot of dimensions (784 dimensions), which are not easy to be reduced. Using euclidean distance to the high dimensional dataset often falls into curse of dimensionality. As a result, it was not even easy to get 70% of ARS(Adjusted RAND Score) at the best K value. Therefore, applying K-Means to the fasion-MINST dataset is easy, but the performance is low. We should find a good way to reduce dimensions for better performance.
